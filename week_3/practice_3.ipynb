```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
img = cv.imread('img/pic_1.png', 0)

plt.subplot(1, 2, 1)
plt.title('Original')
plt.imshow(img)

rows, cols = img.shape

img_rotation = cv.warpAffine(img, cv.getRotationMatrix2D((cols/2, rows/2),30, 0.6), (cols, rows))

plt.subplot(1, 2, 2)
plt.title('Rotated Image')
plt.imshow(img_rotation)

cv.imshow('img', img_rotation)
cv.imwrite('img/rotation_out.png', img_rotation)
cv.waitKey(0)
cv.destroyAllWindows()
```
