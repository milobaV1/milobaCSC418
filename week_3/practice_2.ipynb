```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
img = cv.imread('img/pic_1.png', 0)

plt.subplot(1, 2, 1)
plt.title('Original')
plt.imshow(img)

rows, cols = img.shape
M = np.float32([[1, 0, 0], [0, -1, rows], [0, 0, 1]])
reflected_img = cv.warpPerspective(img, M, (int(cols), int(rows)))

plt.subplot(1, 2, 2)
plt.title('Reflected Image')
plt.imshow(reflected_img)

cv.imwrite('img/reflection_out.png', reflected_img)
cv.waitKey(0)
cv.destroyAllWindows()
```
