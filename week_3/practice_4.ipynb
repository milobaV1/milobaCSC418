```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
img = cv.imread('img/pic_1.png', 0)

plt.subplot(1, 2, 1)
plt.title('Original')
plt.imshow(img)

cropped_img = img[450:700, 200:500]

plt.subplot(1, 2, 2)
plt.title('Cropped Image')
plt.imshow(cropped_img)

cv.imwrite('img/cropped_out.png', cropped_img)
cv.waitKey(0)
cv.destroyAllWindows()
```
