```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv.imread('img/pic_1.png')

plt.subplot(1, 2, 1)
plt.title("Original")
plt.imshow(img)

brightness = 5
contrast = 1.5

img2 = cv.addWeighted(img, contrast, np.zeros(img.shape, img.dtype), 0, brightness)

cv.imwrite('img/contrast_image.png', img2)

plt.subplot(1, 2, 2)
plt.title("Brightness and Contrast")
plt.imshow(img2)
plt.show()
```
