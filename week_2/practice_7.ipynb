```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv.imread('img/pic_1.png')

plt.subplot(1, 2, 1)
plt.title("Original")
plt.imshow(img)

inverse_image = 255 - img

cv.imwrite('img/inverse.png', inverse_image)

plt.subplot(1, 2, 2)
plt.title("Inverse Color")
plt.imshow(inverse_image)
plt.show()
```
