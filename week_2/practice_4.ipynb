```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv.imread('img/pic_1.png')


plt.subplot(1, 2, 1)
plt.title("Original")
plt.imshow(img)


kernel = np.array([[0, -1, 0], [-1, 5, -1], [0, -1, 0]])

sharpened_image = cv.filter2D(img, -1, kernel)

cv.imwrite('img/sharpened_image.png', sharpened_image)

plt.subplot(1,2,2)
plt.title('Sharpening')
plt.imshow(sharpened_image)
plt.show()
```
