```python
import cv2 as cv 
import matplotlib.pyplot as plt
```

```python
img = cv.imread('img/pic_1.png')

plt.subplot(1, 2, 1)
plt.title("Original")
plt.imshow(img)

scaled_image = cv.resize(img, None, fx=2, fy=2)

cv.imwrite('img/scaled.png', scaled_image)

plt.subplot(1, 2, 2)
plt.title("Scaled Image")
plt.imshow(scaled_image)
plt.show()
```
