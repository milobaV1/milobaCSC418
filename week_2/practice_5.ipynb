```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv.imread('img/igho_1.jpg')

plt.subplot(1, 2, 1)
plt.title("Original")
plt.imshow(img)

filtered_image = cv.medianBlur(img, 15)

cv.imwrite('img/median_slur.png', filtered_image)

plt.subplot(1, 2, 2)
plt.title("Median Blur")
plt.imshow(filtered_image)
plt.show()
```
